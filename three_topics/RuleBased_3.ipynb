```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

```

```python


dataframe = pd.read_csv('../data/labelled_dataset_3_topics.csv')
dataframe = dataframe.dropna(how='all')
# can be features other than the text column
features = dataframe["Text"]
labels = dataframe["Best Topic"]
# targets = ["1", "2", "3", "4", "5", "6"]
targets = ["1", "2", "3"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
term_to_topic_df = pd.read_csv('../data/term_to_topic.csv')
term_to_topic_df
```

```python

def predict_single_text(text,idx):
   topic_n_scores = {}
   
   for i in targets:
      topic_n_scores[int(i)] = 0
   # for topic_n_scores
   for word in text.split(' '):
      for term_ntopic_freq in term_to_topic_df.loc[term_to_topic_df['Term']==word].values.tolist():
         topic_n_scores[term_ntopic_freq[1]]+=term_ntopic_freq[2]
   # print(topic_n_scores)
   # print(np.argmax(list(topic_n_scores.values()))+1)
   output = np.argmax(list(topic_n_scores.values()))+1
   # print(f'Predict {output} vs. Actual {list(y_test)[idx]}')
   return output



def predict(X_test)->list:
   output = []
   for i,text in enumerate(list(X_test)):
      output.append(predict_single_text(text, i))
   
   return output
```

```python
predictions = predict(X_test=X_test)
score = accuracy_score(predictions, list(y_test))

print("Rule Based Accuracy Score -> ", score)
```
